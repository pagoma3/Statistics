```python
import pandas as pd
import sqlite3
```

```python
sql = """
SELECT finishing, heading_accuracy, volleys, free_kick_accuracy, shot_power, long_shots, penalties
FROM PLAYER_ATTRIBUTES
"""

con = sqlite3.connect("database.sqlite")
```

```python
attributes = pd.read_sql_query(sql, con)
```

Agafa un conjunt de dades de tema esportiu que t'agradi i selecciona un atribut del conjunt de dades. Calcula la moda, la mediana, la desviació estàndard i la mitjana aritmètica:

```python
attributes["finishing"].mode()
```

```python
attributes["finishing"].median()
```

```python
attributes["finishing"].std()
```

```python
attributes["finishing"].mean()
```

Continuant amb les dades de tema esportiu, selecciona dos atributs i calcula'n la seva correlació.

```python
attributes["finishing"].corr(attributes["penalties"])

# Alta correlació, que té sentit donat que estem mesurant la coincidència de qualitat de xut i llançament de penaltis
```

Continuant amb les dades de tema esportiu, calcula la correlació de tots els atributs entre sí i representa'ls en una matriu amb diferents colors d'intensitat.

```python
import seaborn as sns
import numpy as np
```

```python
corr = attributes.corr()

sns.heatmap(corr, cmap = "flare", center= 1, annot = True)
```

Continuant amb les dades de tema esportiu, selecciona un atribut i calcula la mitjana geomètrica i la mitjana harmònica.

```python
from scipy import stats

attributes.dropna(inplace = True)

stats.hmean(attributes['volleys'])
```
